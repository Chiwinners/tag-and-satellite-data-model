```python
from pathlib import Path
import xarray as xr

# 1) Ruta al directorio (ajústala si tu notebook está en otra carpeta)
base = Path.cwd()
candidates = [
    base / "downloads" / "clorophyll" / "sample",
    base.parent / "downloads" / "clorophyll" / "sample",
    base.parent.parent / "downloads" / "clorophyll" / "sample",
]
sample_dir = next((p.resolve() for p in candidates if p.is_dir()), None)
assert sample_dir is not None, f"No se encontró downloads/clorophyll/sample desde {base}"

# 2) Un archivo .nc
nc_files = sorted(sample_dir.glob("*.nc"))
assert nc_files, f"No hay .nc en {sample_dir}"
nc_path = nc_files[0]  # cámbialo si quieres abrir otro

print(f"Abrir: {nc_path}")

# 3) Elige un engine que tengas instalado: 'netcdf4' (recomendado) o 'h5netcdf'
ds = xr.open_dataset(nc_path, engine="netcdf4", decode_cf=True, mask_and_scale=True)

# 4) Mostrar el Dataset (tú inspeccionas)
ds
```
